```python
categorical = df.select_dtypes(include="object").columns.tolist()
numerical = df.select_dtypes(exclude="object").columns.tolist()

print("Before modification:")
print("Categorical:", categorical)
print("Numerical:", numerical)

# 1. Remove 'SeniorCitizen' from numerical if it exists
if 'SeniorCitizen' in numerical:
    numerical.remove('SeniorCitizen')

# 2. Append 'SeniorCitizen' to categorical if it's not already there
if 'SeniorCitizen' not in categorical:
    categorical.append('SeniorCitizen') 

print("\nAfter modification:")
print("Categorical:", categorical)
print("Numerical:", numerical)
```

```python
import math

cols_per_row = 3
total_plots = len(categorical)
rows = math.ceil(total_plots / cols_per_row)

# Initialize the grid layout
fig, axes = plt.subplots(rows, cols_per_row, figsize=(18, 5 * rows))
axes = axes.flatten()

# Populate the grid dynamically
for idx, col in enumerate(categorical):
    sns.countplot(data=df, x=col, hue="Churn", ax=axes[idx])
    axes[idx].set_title(f"{col} vs Churn", fontsize=12, fontweight='bold')
    axes[idx].tick_params(axis='x', rotation=30)
    axes[idx].set_xlabel(col)
    axes[idx].set_ylabel("Count")

# Remove any unused subplot axes from the grid
for empty_idx in range(idx + 1, len(axes)):
    fig.delaxes(axes[empty_idx])

plt.tight_layout()
plt.show()
```
